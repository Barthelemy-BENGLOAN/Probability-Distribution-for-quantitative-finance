```python
import numpy as np 
import matplotlib.pyplot as plt
```

## Normal (Gaussian) Distribution
PDF:  
$ f(x;\mu,\sigma^2) = \frac{1}{\sigma \sqrt{2\pi}} \exp\!\left(-\frac{(x-\mu)^2}{2\sigma^2}\right), \quad x \in \mathbb{R} $

Parameters: mean $\mu$, variance $\sigma^2$  
Support: $\mathbb{R}$  
Mean = $\mu$, Variance = $\sigma^2$


```python
def normal_pdf(x, mu, sigma2):
    sigma = np.sqrt(sigma2)
    pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2) / (2 * sigma2))
    return pdf
    
x = np.linspace(-5, 5, 1000)

#different parameters
params = [(0, 1.0),(0, 0.2),(0, 5),(-2, 0.5)]

plt.figure(figsize=(8,6))
for mu, sigma2 in params:
    y = normal_pdf(x, mu, sigma2)
    plt.plot(x, y, lw=2, label=fr"$\mu={mu}, \ \sigma^2={sigma2}$")

plt.title("Normal Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

## Log-Normal Distribution
PDF:  
$ f(x;\mu,\sigma^2) = \frac{1}{x\sigma\sqrt{2\pi}} \exp\!\left(-\frac{(\ln x - \mu)^2}{2\sigma^2}\right), \quad x > 0 $

Parameters: $\mu$, $\sigma > 0$  
Support: $x > 0$

```python
def log_normal_pdf(x, mu, sigma2):
    sigma = np.sqrt(sigma2)
    pdf = (1 / (x * sigma * np.sqrt(2 * np.pi))) * np.exp(-((np.log(x) - mu) ** 2) / (2 * sigma2))
    return pdf

x=np.linspace(1e-3,2.5,1500)

param=[(0,1),(0,0.25),(0,0.5),(0,2),(-1,0.5)]

plt.figure(figsize=(8,6))
for mu, sigma2 in param:
    y = log_normal_pdf(x, mu, sigma2)
    plt.plot(x, y, lw=2, label=fr"$\mu={mu}, \ \sigma^2={sigma2}$")

plt.title("Log-Normal Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

## Inverse Gaussian (Wald distribution)
PDF:  
$ f(x;\mu,\lambda) = \sqrt{\frac{\lambda}{2\pi x^3}} \,\exp\!\left( -\frac{\lambda(x-\mu)^2}{2\mu^2 x} \right), \quad x > 0 $

Parameters: mean $\mu > 0$, shape $\lambda > 0$


```python
def inverse_normal_pdf(x, mu, lam):
    pdf = np.sqrt(lam/(2*np.pi*x**3))*np.exp(-lam*(x-mu)**2/(2*mu**2*x))
    return pdf

x=np.linspace(1e-3,2.5,1000)

param=[(1,0.2),(1,1),(1,3),(3,0.2),(3,1)]

plt.figure(figsize=(8,6))
for mu, lam in param:
    y = inverse_normal_pdf(x, mu, lam)
    plt.plot(x, y, lw=2, label=fr"$\mu={mu}, \ \lambda={lam}$")

plt.title(" Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

## Fréchet Distribution
PDF:  
$ f(x;\alpha,s) = \frac{\alpha}{s} \left(\frac{s}{x}\right)^{1+\alpha} \exp\!\left[-\left(\frac{s}{x}\right)^\alpha\right], \quad x > 0 $

Parameters: shape $\alpha > 0$, scale $s > 0$

```python
def frechet_pdf(x, alpha, s, m=0.0):
    z = (x - m) / s
    return (alpha / s) * z**(-(1 + alpha)) * np.exp(- z**(-alpha))

x = np.linspace(1e-3, 5.0, 1500)
param = [(1, 1, 0),(2, 1, 0),(3, 1, 0),(1, 2, 0),(2, 2, 0),(3, 2, 0)]

plt.figure(figsize=(8,6))
for alpha, s, m in param:
    y = frechet_pdf(x, alpha, s, m)
    plt.plot(x, y, lw=2, label=fr"$\alpha={alpha};\ s={s};\ m={m}$")

plt.title("Fréchet Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.xlim(0, 5)
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

## Weibull Distribution
PDF:  
$ f(x;k,\lambda) = \frac{k}{\lambda} \left(\frac{x}{\lambda}\right)^{k-1} e^{-(x/\lambda)^k}, \quad x \geq 0 $

Parameters: shape $k > 0$, scale $\lambda > 0$

```python
def weibull_pdf(x, k, lam):
    return (k/lam) * (x/lam)**(k-1) * np.exp(-(x/lam)**k)

x = np.linspace(0.01, 5.0, 1000)
params = [(1.0, 0.5),(1.0, 1.0),(1.0, 1.5),(1.0, 5.0)]

plt.figure(figsize=(8,6))
for lam, k in params:
    y = weibull_pdf(x, k, lam)
    plt.plot(x, y, lw=2, label=fr"$\lambda={lam},\ k={k}$")

plt.title("Weibull Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.xlim(0, 2.5)
plt.ylim(0,2.5)
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

## Gumbel Distribution
PDF:  
$ f(x;\mu,\beta) = \frac{1}{\beta} \exp\!\left(-\frac{x-\mu}{\beta} - e^{-(x-\mu)/\beta}\right), \quad x \in \mathbb{R} $

Parameters: location $\mu$, scale $\beta > 0$


```python
def gumbel_pdf(x, mu, beta):
    return (1/beta) * np.exp(-(x - mu)/beta - np.exp(-(x - mu)/beta))

x = np.linspace(-5, 25, 1000)
param = [(0.5, 2),(1, 2),(1.5, 3),(3, 4)]

plt.figure(figsize=(8,6))
for mu, beta in param:
    y = gumbel_pdf(x, mu, beta)
    plt.plot(x, y, lw=2, label=fr"$\mu={mu},\ k={beta}$")

plt.title("Gumbel Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

## Exponential Distribution
PDF:  
$ f(x;\lambda) = \lambda e^{-\lambda x}, \quad x \geq 0 $

Parameter: rate $\lambda > 0$  
Mean = $1/\lambda$, Variance = $1/\lambda^2$

```python
def exponential_pdf(x, lam):
    return lam * np.exp(-lam * x)

x = np.linspace(0, 5, 1000)
param = [0.5,1,1.5]

plt.figure(figsize=(8,6))
for lam in param:
    y = exponential_pdf(x, lam)
    plt.plot(x, y, lw=2, label=fr"$\lambda={lam}$")

plt.title("Exponential Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

## Gamma Distribution
PDF:  
$ f(x;k,\theta) = \frac{1}{\Gamma(k)\theta^k} x^{k-1} e^{-x/\theta}, \quad x > 0 $

Parameters: shape $k > 0$, scale $\theta > 0$


```python
from math import gamma as gm

def gamma_pdf(x, k, theta):
    return (1 / (gm(k) * theta**k)) * x**(k - 1) * np.exp(-x / theta)

x = np.linspace(0, 20, 1000)
param=[(1,2),(2,2),(3,2),(5,1),(9,0.5)]

plt.figure(figsize=(8,6))
for k, theta in param:
    y = gamma_pdf(x, k, theta)
    plt.plot(x, y, lw=2, label=fr"$ k={k},\ \theta={theta}$")

plt.title("Gamma Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(alpha=0.3)
plt.legend()
plt.show()
```

## $\chi^2$ Distribution
PDF:  
$ f(x;k) = \frac{1}{2^{k/2}\Gamma(k/2)} x^{k/2 -1} e^{-x/2}, \quad x > 0 $

Parameter: degrees of freedom $k$  
Mean = $k$, Variance = $2k$


```python
def chisquare_pdf(x, k):
    pdf= (1/(2**(k/2)*gm(k/2)))*x**((k/2)-1)*np.exp(-x/2)
    return pdf
    
x = np.linspace(1e-3, 10, 1000)
param = [1, 2, 3, 4, 6, 9]

plt.figure(figsize=(8,6))
for k in param:
    y = chisquare_pdf(x, k)
    plt.plot(x, y, lw=2, label=fr"$k={k}$")

plt.title("Chi-Square Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(alpha=0.3)
plt.ylim(0,0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

## Student’s t Distribution
PDF:  
$ f(x;\nu) = \frac{\Gamma\left(\frac{\nu+1}{2}\right)}{\sqrt{\nu\pi}\,\Gamma(\nu/2)} \left(1+\frac{x^2}{\nu}\right)^{-\frac{\nu+1}{2}}, \quad x \in \mathbb{R} $

Parameter: degrees of freedom $\nu > 0$


```python
def student_t_pdf(x, nu):
    pdf = (gm((nu + 1) / 2.0) / (np.sqrt(nu * np.pi) * gm(nu / 2.0)))* (1.0 + (x**2) / nu) ** (-(nu + 1) / 2.0)
    return pdf

x = np.linspace(-10, 10, 1000)
param = [1, 2, 5]

plt.figure(figsize=(8,6))
for nu in param:
    y = student_t_pdf(x, nu)
    plt.plot(x, y, lw=2, label=fr"$\nu={nu}$")

plt.title("Student t Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(alpha=0.3)
plt.ylim(0, 0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

## Cauchy Distribution
PDF:  
$ f(x;x_0,\gamma) = \frac{1}{\pi\gamma \left[1+\left(\frac{x-x_0}{\gamma}\right)^2\right]}, \quad x \in \mathbb{R} $

Parameters: location $x_0$, scale $\gamma > 0$  
(No mean, no variance)


```python
def cauchy_pdf(x, x0, gamma):
    pdf = 1 / (np.pi * gamma * (1 + ((x - x0) / gamma) ** 2))
    return pdf

x = np.linspace(-5, 5, 1000)
param = [(0,0.5), (0,1), (0,2), (-2,1)]

plt.figure(figsize=(8,6))
for x0, gam in param:
    y = cauchy_pdf(x, x0, gam)
    plt.plot(x, y, lw=2, label=fr"$ x0={x0},\ \gamma={gam}$")

plt.title("Cauchy Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(alpha=0.3)
plt.ylim(0, 0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

## Beta Distribution
PDF:  
$ f(x;\alpha,\beta) = \frac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha,\beta)}, \quad 0 < x < 1 $

Parameters: $\alpha > 0$, $\beta > 0$

```python
def betaf(alpha, beta):
    return gm(alpha) * gm(beta) / gm(alpha + beta)

def beta_pdf(x,alpha,beta):
    pdf = (x**(alpha - 1)) * ((1 - x)**(beta - 1)) / betaf(alpha, beta)
    return pdf
    
x=np.linspace(1e-3,0.999,1000)
param=[(0.5,0.5), (5,1), (1,3), (2,2), (2,5)]

for alpha, beta in param:
    y = beta_pdf(x, alpha, beta)
    plt.plot(x, y, lw=2, label=fr"$ \alpha={alpha},\ \beta={beta}$")

plt.title("Beta Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(alpha=0.3)
plt.ylim(0, 3)
plt.legend()
plt.tight_layout()
plt.show()
```

## Laplace Distribution
PDF:  
$ f(x;\mu,b) = \frac{1}{2b} \exp\!\left(-\frac{|x-\mu|}{b}\right), \quad x \in \mathbb{R} $

Parameters: location $\mu$, scale $b > 0$

```python
def laplace_pdf(x, mu, b):
    pdf = 1/(2*b)*np.exp(-np.abs(x-mu)/b)
    return pdf

x = np.linspace(-10, 10, 1000)
param = [(0,1), (0,2), (0,4), (-5,4)]

plt.figure(figsize=(8,6))
for mu, b in param:
    y = laplace_pdf(x, mu, b)
    plt.plot(x, y, lw=2, label=fr"$ \mu={mu},\ b={b}$")

plt.title("Laplace Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(alpha=0.3)
plt.ylim(0, 0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

## Lévy Distribution
PDF:  
$ f(x;\mu,c) = \sqrt{\frac{c}{2\pi}} \frac{e^{-c/(2(x-\mu))}}{(x-\mu)^{3/2}}, \quad x > \mu $

Parameters: location $\mu$, scale $c > 0$

```python
def levy_pdf(x, mu, c):
    pdf = (np.sqrt(c/(2*np.pi))*np.exp(-c/(2*(x-mu))))/(x-mu)**(3/2)
    return pdf

x = np.linspace(1e-3, 5, 1000)
param = [(0,1/2), (0,1), (0,2), (0,4), (0,8)]

plt.figure(figsize=(8,6))
for mu, c in param:
    y = levy_pdf(x, mu, c)
    plt.plot(x, y, lw=2, label=fr"$ \mu={mu},\ c={c}$")

plt.title("Lévy Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(alpha=0.3)
plt.ylim(0, 1)
plt.legend()
plt.tight_layout()
plt.show()
```

## Logistic Distribution
PDF:  
$ f(x;\mu,s) = \frac{e^{-(x-\mu)/s}}{s \left(1+e^{-(x-\mu)/s}\right)^2}, \quad x \in \mathbb{R} $

Parameters: location $\mu$, scale $s > 0$

```python
def logistic_pdf(x, mu, s):
    pdf = np.exp(-(x-mu)/s)/(s*(1+np.exp(-(x-mu)/s))**2)
    return pdf

x = np.linspace(-5, 20, 1000)
param = [(5,2), (9,3), (9,4), (6,2), (2,1)]

plt.figure(figsize=(8,6))
for mu, s in param:
    y = logistic_pdf(x, mu, s)
    plt.plot(x, y, lw=2, label=fr"$ \mu={mu},\ c={s}$")

plt.title("Logistic Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(alpha=0.3)
plt.ylim(0, 0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

## Uniform Distribution
PDF:  
$ f(x;a,b) = \frac{1}{b-a}, \quad a \leq x \leq b $

Parameters: $a < b$



```python
def uniform_pdf(x, a, b):
    pdf = np.where((x >= a) & (x <= b), 1.0/(b - a), 0.0)
    return pdf

x = np.linspace(-2, 8, 1000)
param = [(0,1), (2,5), (-1,3)]

plt.figure(figsize=(8,6))
for a, b in param:
    y = uniform_pdf(x, a, b)
    plt.plot(x, y, lw=2, label=fr"$a={a},\ b={b}$")

plt.title("Uniform Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(alpha=0.3)
plt.ylim(0, 1.2)
plt.legend()
plt.tight_layout()
plt.show()
```

## Pareto
PDF:  
$ f(x;x_m,\alpha) = \frac{\alpha x_m^\alpha}{x^{\alpha+1}}, \quad x \geq x_m > 0 $

Parameters: scale $x_m > 0$, shape $\alpha > 0$

```python
def pareto_pdf(x, xm, alpha):
    pdf = np.where(x >= xm, alpha * xm**alpha / (x**(alpha + 1)), 0.0)
    return pdf

x = np.linspace(0.1, 10, 1000)
param = [(1, 1), (1, 2), (2, 3)]

plt.figure(figsize=(8,6))
for xm, alpha in param:
    y = pareto_pdf(x, xm, alpha)
    plt.plot(x, y, lw=2, label=fr"$x_m={xm},\ \alpha={alpha}$")

plt.title("Pareto Distribution PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.grid(alpha=0.3)
plt.ylim(0, 2.1)
plt.legend()
plt.tight_layout()
plt.show()
```
